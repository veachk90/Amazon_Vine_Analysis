```python
import pandas as pd

vine_df = pd.read_csv('../Module_17/Vine_Review_Analysis.csv')
vine_df.head(10)
```

```python
total_votes_df = vine_df[vine_df['total_votes'] >= 20]
total_votes_df.head(10)
```

```python
helpful_votes_df = total_votes_df[total_votes_df['helpful_votes'] / total_votes_df['total_votes'] >= .5]
helpful_votes_df.head(10)
```

```python
paid_reviews_df = helpful_votes_df[helpful_votes_df['vine'] == 'Y']
paid_reviews_df.head(10)
```

```python
unpaid_reviews_df = helpful_votes_df[helpful_votes_df['vine'] == 'N']
unpaid_reviews_df.head(10)
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
# for paid reviews.
paid_count = paid_reviews_df.count()
# print(paid_count)
paid_top_reviews = len(paid_reviews_df[paid_reviews_df['star_rating'] >= 5])
# print(paid_top_reviews)
paid_reviews_percentage = paid_top_reviews / paid_count
print(paid_reviews_percentage)
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
# for unpaid reviews.
unpaid_count = unpaid_reviews_df.count()
unpaid_top_reviews = len(unpaid_reviews_df[unpaid_reviews_df['star_rating'] >= 5])
unpaid_reviews_percentage = unpaid_top_reviews / unpaid_count
print(unpaid_reviews_percentage)
```
